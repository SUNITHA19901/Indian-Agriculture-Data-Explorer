```python
def get_crop_data(df, crop_name):
    """
    Extracts and cleans area and production data for a given crop.
    
    Parameters:
        df (pd.DataFrame): Original dataframe
        crop_name (str): Name of the crop in uppercase (e.g., 'RICE', 'MAIZE')
    
    Returns:
        pd.DataFrame: Cleaned DataFrame with columns ['Area', 'Production']
    """
    
    area_col = f"{crop_name} AREA (1000 ha)"
    prod_col = f"{crop_name} PRODUCTION (1000 tons)"

    
    df.columns = df.columns.str.strip()

    
    if area_col not in df.columns or prod_col not in df.columns:
        raise ValueError(f"Columns for '{crop_name}' not found in the dataset.")

   
    crop_df = df[[area_col, prod_col]].copy()
    crop_df.rename(columns={
        area_col: 'Area',
        prod_col: 'Production'
    }, inplace=True)

    
    crop_df.dropna(subset=['Area', 'Production'], inplace=True)

    return crop_df
```

```python
import pandas as pd
import sqlite3


df_raw = pd.read_csv('D:\AgriData\env\Scripts\cleaned_dataset.csv')


crops = {
    'Rice': ['RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)', 'RICE YIELD (Kg per ha)'],
    'Wheat': ['WHEAT AREA (1000 ha)', 'WHEAT PRODUCTION (1000 tons)', 'WHEAT YIELD (Kg per ha)'],
    'Maize': ['MAIZE AREA (1000 ha)', 'MAIZE PRODUCTION (1000 tons)', 'MAIZE YIELD (Kg per ha)'],
    'Cotton': ['COTTON AREA (1000 ha)', 'COTTON PRODUCTION (1000 tons)', 'COTTON YIELD (Kg per ha)'],
    'Groundnut': ['GROUNDNUT AREA (1000 ha)', 'GROUNDNUT PRODUCTION (1000 tons)', 'GROUNDNUT YIELD (Kg per ha)'],
    'Oilseeds': ['OILSEEDS AREA (1000 ha)', 'OILSEEDS PRODUCTION (1000 tons)', 'OILSEEDS YIELD (Kg per ha)']
}


records = []

for crop, cols in crops.items():
    for _, row in df_raw.iterrows():
        if pd.notna(row[cols[0]]) and pd.notna(row[cols[1]]):
            records.append({
                'Year': row['Year'],
                'State': row['State Name'],
                'District': row['Dist Name'],
                'Crop': crop,
                'Area': row[cols[0]],
                'Production': row[cols[1]],
                'Yield': row[cols[2]]
            })

df_normalized = pd.DataFrame(records)
```

```python

conn = sqlite3.connect('agriculture.db')

query_1 = """
WITH Top3States AS (
    SELECT [State Name]
    FROM agri_data
    GROUP BY [State Name]
    ORDER BY SUM([RICE PRODUCTION (1000 tons)]) DESC
    LIMIT 3
)
SELECT Year, [State Name] AS State,
       SUM([RICE PRODUCTION (1000 tons)]) AS Yearly_Rice_Production
FROM agri_data
WHERE [State Name] IN (SELECT [State Name] FROM Top3States)
GROUP BY Year, [State Name]
ORDER BY Year, State;
"""

df_1 = pd.read_sql_query(query_1, conn)
df_1.to_csv('Q1_Top3_Rice_Production_Yearwise.csv', index=False)


# 2. Top 5 Districts by Wheat Yield Increase Over Last 5 Years
query_2 = """
SELECT [Dist Name], 
       MAX(Year) - MIN(Year) AS Year_Span,
       MAX([WHEAT YIELD (Kg per ha)]) - MIN([WHEAT YIELD (Kg per ha)]) AS Yield_Increase
FROM agri_data
WHERE Year >= (SELECT MAX(Year)-4 FROM agri_data)
GROUP BY [Dist Name]
ORDER BY Yield_Increase DESC
LIMIT 5;
"""
df_2 = pd.read_sql_query(query_2, conn)
df_2.to_csv('Q2_Wheat_Yield_Top5_Districts.csv', index=False)

# 3. States with Highest Growth in Oilseed Production (5-Year Growth)
query_3 = """
WITH prod_by_year AS (
    SELECT [State Name], Year, SUM([OILSEEDS PRODUCTION (1000 tons)]) AS oilseed_prod
    FROM agri_data
    GROUP BY [State Name], Year
),
growth AS (
    SELECT a.[State Name],
           (b.oilseed_prod - a.oilseed_prod) * 100.0 / NULLIF(a.oilseed_prod, 0) AS Growth_Percent
    FROM prod_by_year a
    JOIN prod_by_year b ON a.[State Name] = b.[State Name]
    WHERE b.Year = (SELECT MAX(Year) FROM agri_data)
      AND a.Year = (SELECT MAX(Year)-4 FROM agri_data)
)
SELECT [State Name], ROUND(Growth_Percent, 2) AS Growth_Rate
FROM growth
ORDER BY Growth_Rate DESC
LIMIT 5;
"""
df_3 = pd.read_sql_query(query_3, conn)
df_3.to_csv('Q3_Oilseed_Production_Growth.csv', index=False)

# 4. District-wise Correlation Between Area and Production for Rice, Wheat, Maize
query_4 = """
SELECT [Dist Name], 
       [RICE AREA (1000 ha)], [RICE PRODUCTION (1000 tons)],
       [WHEAT AREA (1000 ha)], [WHEAT PRODUCTION (1000 tons)],
       [MAIZE AREA (1000 ha)], [MAIZE PRODUCTION (1000 tons)]
FROM agri_data;
"""
df_4 = pd.read_sql_query(query_4, conn)
df_4.to_csv('Q4_Crop_Correlation.csv', index=False)

# 5. Yearly Cotton Production Growth in Top 5 States
query_5 = """
SELECT [State Name], Year, SUM([COTTON PRODUCTION (1000 tons)]) AS Total_Cotton_Production
FROM agri_data
GROUP BY [State Name], Year
HAVING [State Name] IN (
    SELECT [State Name]
    FROM agri_data
    GROUP BY [State Name]
    ORDER BY SUM([COTTON PRODUCTION (1000 tons)]) DESC
    LIMIT 5
);
"""
df_5 = pd.read_sql_query(query_5, conn)
df_5.to_csv('Q5_Cotton_Production_Top5_States.csv', index=False)

# 6. Districts with the Highest Groundnut Production in 2020
query_6 = """
SELECT [Dist Name], [State Name], [GROUNDNUT PRODUCTION (1000 tons)]
FROM agri_data
WHERE Year = 2020
  AND [GROUNDNUT PRODUCTION (1000 tons)] IS NOT NULL
  AND [GROUNDNUT PRODUCTION (1000 tons)] > 0
ORDER BY [GROUNDNUT PRODUCTION (1000 tons)] DESC
LIMIT 10;
"""

# Execute the query and save to CSV
df_6 = pd.read_sql_query(query_6, conn)
df_6.to_csv('Q6_Groundnut_Top_Districts_2020.csv', index=False)



# 7. Annual Average Maize Yield Across All States
query_7 = """
SELECT Year, AVG([MAIZE YIELD (Kg per ha)]) AS Avg_Maize_Yield
FROM agri_data
GROUP BY Year;
"""
df_7 = pd.read_sql_query(query_7, conn)
df_7.to_csv('Q7_Avg_Maize_Yield.csv', index=False)

# 8. Total Area Cultivated for Oilseeds by State
query_8 = """
SELECT [State Name], Year, SUM([OILSEEDS AREA (1000 ha)]) AS Total_Oilseed_Area
FROM agri_data
GROUP BY [State Name], Year
ORDER BY [State Name], Year;
"""
df_8 = pd.read_sql_query(query_8, conn)
df_8.to_csv('Q8_Oilseed_Area_By_State_And_Year.csv', index=False)


# 9. Districts with Highest Rice Yield
query_9 = """
SELECT [Dist Name], [State Name], MAX([RICE YIELD (Kg per ha)]) AS Max_Rice_Yield
FROM agri_data
GROUP BY [Dist Name]
ORDER BY Max_Rice_Yield DESC
LIMIT 10;
"""
df_9 = pd.read_sql_query(query_9, conn)
df_9.to_csv('Q9_Top_Rice_Yield_Districts.csv', index=False)

# 10. Compare Wheat and Rice Production in Top 5 States Over 10 Years
query_10 = """
-- Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years
WITH Top5States AS (
    SELECT [State Name]
    FROM agri_data
    WHERE Year >= 2010
    GROUP BY [State Name]
    ORDER BY SUM([WHEAT PRODUCTION (1000 tons)] + [RICE PRODUCTION (1000 tons)]) DESC
    LIMIT 5
)
SELECT [State Name], Year,
       SUM([WHEAT PRODUCTION (1000 tons)]) AS Wheat_Prod,
       SUM([RICE PRODUCTION (1000 tons)]) AS Rice_Prod
FROM agri_data
WHERE Year >= 2010 AND [State Name] IN (SELECT [State Name] FROM Top5States)
GROUP BY [State Name], Year
ORDER BY Year, [State Name];
"""


df_10 = pd.read_sql_query(query_10, conn)


df_10.to_csv('Q10_Wheat_Rice_Top5_States_2010s.csv', index=False)






conn.close()
```
